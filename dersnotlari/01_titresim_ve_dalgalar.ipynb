# Basit Sistemlerin Serbest Salınımları

## Giriş

- Evren ve içindeki her şey sürekli hareket halindedir.
- Hareketli cisimler iki ana sınıfa ayrılabilirler.
    - Başladıkları konumlara dönenler:
        - salınan bir sarkaç
        - titreşen bir keman teli
        - bardakta ileri geri çalkalanan su
        - atom çekirdeği etrafında dolanan elektronlar
        - aynalar arasında gidip gelen lazer ışını
    - İlk başladıkları konumdan farklı bir konuma gidenler:
        - buzlu yüzeyde kayan bir cisim
        - bir ipte ilerleyen atma (puls)
        - denizde kıyıya doğru ilerleyen dalgalar
        - elektron tabancasından çıkıp uçan elektron
        - bir yıldızdan gözümüze gelen ışık
- Bazen aynı olay her iki anasınıfta da düşünülebilir.
    - Deniz dalgaları ilerler fakat su dalgası üzerindeki bir cisim ilerlemez, bulunduğu konumda yukarı aşağı iner çıkar.
    - İpteki atmalar ilerler ama ipin atomları bir denge noktası etrafında salınırlar.
- Bu bölümde dışarıdan bir kere uyarıldıktan sonra bir denge konumu etrafında serbestçe salınan sistemleri inceleyeceğiz.
- Bu tür salınımlara **serbest salınımlar** veya **doğal salınımlar** denir.
- Birinci bölümde bir veya iki cisim içeren sistemleri inceleyeceğiz.
- İkinci bölümde ise çok serbestlik dereceli sistemleri inceleyeceğiz. 
- Çok serbestlik dereceli bu sistemlerin **kip** (mod, öztitreşim) olarak adlandırılan bir serbestlik dereceli salınıcıya benzer yalın hareketlerinin olduğunu göreceğiz.
- Bu tür sistemlerin genel davranışlarının bu yalın hareketlerlerin birleşimleriyle ifade edilebileceğiniz göreceğiz.
- İncelenecek sistemler çeşitli fiziksel nicelikler olabilir ve dalga fonksiyonları aşağıdaki gibi gösterilirler.
    - Mekanik bir sistemse genellikle, $\Psi(x,y,zt)$
    - Elektromanyetik bir sistemse, elektrik alan $E(x,y,z,t)$ ve manyetik alan $B(x,y,zt)$

## Bir serbestlik dereceli sistemlerin serbest salınımları

Yukarıdaki şekilde verilen **kütle-yay** ve **ip-cisim** sistemleri bir serbestlik dereceli salınıcılardır. Çünkü denge konumundan olan uzaklıklarını sadece bir koordinatla ifade etmek mümkündür. Bu tür sistemlerin hareketi Denklem \ref{eq:hs}'deki "harmonik salınım" fonksiyonu ile tanımlanabilir. Şekildeki ***LC* devresi** de bir serbestlik dereceli salınıcıdır.

\begin{equation}
\label{eq:hs}
\large
\Psi(t) = A\sin(\omega t+\phi)
\end{equation}

<img width=300 src='./sekiller/LC_circuit.gif'>


**Geri çağırma ve eylemsizlik**:  
Denklem \ref{eq:hs} ile tanımlanan davranış bu tür sistemlerin iki özelliğinin yarışması sonucu ortaya çıkar.

- geri çağırma kuvveti
- eylemsizlik

**Geri çağırma kuvveti** sistemi hep denge konumuna getirmek ister fakat denge konumuna geldiğinde sistem en yüksek hızına ulaşır. Bu noktada **eylemsizlik** (Newton'un birinci yasası) etkin hale gelir ve sistem hareketine devam etmek ister. Böylece sürekli bir **salınım hareketi** gerçekleşir.

**Sönümlü salınımlar**:  
Denklem \ref{eq:hs} ile hareketi tanımlanan sistemler çoğu durumda gerçekçi olmayabilir. Çünkü çoğu fizikzel durumda salınımı **sönümleyecek** (sonlandıracak) kuvvetler vardır ve bunlar **sürtünme** veya **direnç** kuvvetleri olarak adlandırılırlar. Böyle sistemler için ise salınım fonksiyonu aşağıdaki gibidir.

\begin{equation}
\label{eq:shs}
\large
\Psi(t) = A e^{-t/2\tau} \cos(\omega t+\phi)
\end{equation}

$\bf \tau$ sönme veya durulma zamanı olarak adlandırılır.



### Harmonik salınım fonksiyonu

Harmonik salınım fonksiyonuna ait parametrelerin değişimilerinin fonksiyonda ne tür değişikliklere sebep olduğunu aşağıdaki Python programını çalıştırarak inceleyebilirsiniz.

Programı çalıştırmak için hücrenin kenarındaki **In[ ]** yazan kısma  tıklayın. Sonra **Shit + Enter** tuşlarını beraber kullanarak  çalıştırabilirsiniz.

```python
#*************************************************************
# Gerekli kütüphaneler en üstteki Python hücresinde yüklendi.
#*************************************************************

# Harmonik salınım fonksiyonu
def psi0(A, w, phi):
    return A * np.sin(w*t + phi)

# Zaman uzayını oluştur
t = np.linspace(0, 4*np.pi, 75)

# Fonksiyonun başlangıç hali
y0 = psi0(1, 1, 0.)

# Etkileşimli kısmı başlat
@interact(
    A = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    w = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    phi = FloatSlider(min=-np.pi, max=np.pi, step=np.pi/20, value=0.0, description='$\phi$')
) 

def wave0(A, w, phi):
    plt.suptitle(r'$\Psi(t) = A\sin(wt+\phi),\;\;\phi = \frac{%s}{20} \pi$'%int(20*phi/(np.pi)), fontsize=18, y=1.0)
    plt.grid()
    plt.plot(t, psi0(A, w, phi),'b', lw=3)
    plt.plot(t, y0, 'r--', lw=3)
    plt.xlabel('t', fontsize=18)
    plt.ylabel('$\Psi(t)$', fontsize=18)
    plt.ylim(-2,2)
    plt.show()
```

### Sönümlü salınım fonksiyonu

Sönümlü salınım fonksiyonuna ait parametrelerin değişimilerinin fonksiyonda ne tür değişikliklere sebep olduğunu aşağıdaki Python programını çalıştırarak inceleyebilirsiniz.

Programı çalıştırmak için hücrenin kenarındaki **In[ ]** yazan kısma  tıklayın. Sonra **Shit + Enter** tuşlarını beraber kullanarak  çalıştırabilirsiniz.

```python
#*************************************************************
# Gerekli kütüphaneler en üstteki Python hücresinde yüklendi.
#*************************************************************

# Harmonik salınım fonksiyonu
def psi1(A, w, tau, phi):
    return A * np.exp(-t/2/tau)*np.cos(w*t + phi)

# Zaman uzayını oluştur
t = np.linspace(0, 4*np.pi, 75)

# Fonksiyonun başlangıç hali
y1 = psi1(1, 1, 2.77, 0.)

# Etkileşimli kısmı başlat
@interact(
    A = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    w = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    tau = FloatSlider(min=.1, max=80.1, step=80./30, value=2.77),
    phi = FloatSlider(min=-np.pi, max=np.pi, step=np.pi/20, value=0.0, description='$\phi$')
) 

def wave1(A, w, tau, phi):
    plt.suptitle(r'$\Psi(t) = A e^{-t/2\tau} \cos(wt+\phi),\;\;\phi = \frac{%s}{20} \pi$'%int(20*phi/(np.pi)), fontsize=18, y=1.0)
    plt.grid()
    plt.plot(t, psi1(A, w, tau, phi),'b', lw=3)
    plt.plot(t, y1, 'r--', lw=3)
    plt.xlabel('t', fontsize=18)
    plt.ylabel('$\Psi(t)$', fontsize=18)
    plt.ylim(-2,2)
    plt.show()
```

Kütle-yay sistemi bir serbestlik dereceli sistemlere verilebilecek en kolay fakat en önemli örneklerden birisidir. Kütle-yay sistemi basit harmonik hareket (BHH) davranışı gösterir. Sistem x=0'daki denge konumundan bir kere uzaklaştırıldığında (eğer dışarıdan etkiyle durdurulmazsa) sonsuza kadar denge noktasına gider ve gelir. Bu hareket BHH olarak adlandırılır. Şekildeki gibi bir mekanik sistemde BHH'nin gerçekleşmesi için bir kaç ön kabul yapmak gerekir.
<br>    
<li>Yay kütlesizdir veya yayın kütlesi cismin kütlesi yanında çok küçüktür.</li>
<li>Cisim ve hareket ettiği yüzey arasında sürtünme yoktur.</li>
<li>Yay Hooke yasasına uygun davranır: yay üzerinde oluşan kuvvet sıkıştırılma veya genleştirilme miktarıyla doğru orantılıdır.</li>    
<br>
Bu kabuller altında yay tarafından kütleye uygulanan kuvvet herhangi bir anda,

\begin{equation}
\label{eq:yay}
\large
\vec F_{yay} = - k \vec x
\end{equation}
<br>
ile verilir, burada **k** yay (kuvvet) sabitidir ve birimi Newton/metre (N/m)'dir. Cismin üzerindeki hareketi sağlayan tek kuvvet yay kuvveti olduğundan, Newtonun ikinci yasası uygulanırsa,

\begin{equation}
\label{eq:yayNewton}
\large
\Sigma \vec F = m \vec a \\
\large
\Sigma \vec F = \vec F_{yay}  \\
\large
 m \vec a = - k \vec x
\end{equation}
<br>
elde edilir. $\vec a$ ivme vektörünün, konum vektörü $\vec x$ zamana göre ikinci türevi olduğu hatırlanırsa ve denklem yeniden düzenlenirse,

\begin{equation}
\label{eq:yayDiff}
\large
\frac{d^2\vec x}{dt^2} + \frac{k}{m}\vec x = 0
 \end{equation}
<br> 
ikinci dereceden lineer bir diferansiyel denklem elde edilir. Bu denklem aşağıdaki formda da yazılabilir. Bu tür denklemler BHH diferansiyel denklemleri olarak adlandırılır.
<br>
\begin{equation}
\label{eq:yayDiff2}
\large
\ddot{\vec x} + \omega^2\vec x = 0
\end{equation}
<br> 
Burada $\ddot{\vec x} \equiv \frac{d^2\vec x}{dt^2}$  ve $\omega^2 \equiv {k \over m}$ 'dir. $\omega^2$ birim kütle ve birim uzunluk başına geri çağırıcı kuvvettir. Basit bir incelemeyle,

\begin{equation}
\label{eq:omega}
\large
\frac{F_{yay}}{m x} = k/m = \omega^2
\end{equation}
<br>
olduğu görülür. Bir çok ders kitabında Denk. \ref{eq:yayDiff2}'nin aşağıdaki formda yazıldığı görülür. $\ddot{\vec x}$ ve $\vec x$ aynı <b>doğrultuya</b> 
sahip olduklarından bu notasyonun kullanılmasında bir sakınca yoktur, fakat bu notasyonda bile her ikisininde halen birer zıt yönlü vektör olduklarının unutulmaması gerekir.
<br>
\begin{equation}
\label{eq:yayDiff3}
\large
\ddot{x} + \omega^2 x = 0
\end{equation}
<br>
Bu son halini alan BHH diferansiyel denkleminin genel çözümü aşağıdaki gibidir.
<br><br>
\begin{equation}
\label{eq:yayCozum}
\large
x(t) = a \cos(\omega t) + b \sin(\omega t)
\end{equation}
<br>
$a = A \cos(\phi)$ ve $b = -A\sin(\phi)$ seçilirse çözüm,
<br><br>
\begin{equation}
\label{eq:yayCozum1}
\large
x(t) = A \cos(\omega t + \phi)
\end{equation}
<br>
halini alır. $a = A \sin(\phi)$ ve $b = A\cos(\phi)$ seçilirse çözüm,
<br><br>
\begin{equation}
\label{eq:yayCozum2}
\large
x(t) = A \sin(\omega t + \phi)
\end{equation}
<br>
halini alır. Bu üç çözüm de BHH denkleminin geçerli çözümleridir. Burada $A$ kütle-yay sisteminde kütlenin denge konumuna göre yer değiştirmesinin en büyük olduğu konumları belirler ($x = \pm A$). $\bf \phi$ ise ***faz farkı*** olarak adlandırılır. **A** ve $\omega$ değerleri özdeş iki kütle-yay sistemi aynı anda aynı konumda bulunmaya bilirler, **faz farkı** bu farklılığı hesaba katar. Hem **A** hem de $\bf \phi$, seçilen ve bilinen bir başlangıç değeri için BHH diferansiyel denkleminin genel çözümünün kullanılmasıyla belirlenirler. $\bf \omega$ sistemin açısal frekansıdır ve hem **A** hem de $\bf \phi$'den farklı olarak sistemin başlangıç şartlarından bağımsızdır.
```python
#*************************************************************
# Gerekli kütüphaneler en üstteki Python hücresinde yüklendi.
#*************************************************************

# Harmonik salınım fonksiyonu
def psi2(A, w, phi):
    return A * np.cos(w*t + phi)

def dpsi2(A, w, phi):
    return -w * A * np.sin(w*t + phi)

def ddpsi2(A, w, phi):
    return -w**2 * A * np.cos(w*t + phi)

# Zaman uzayını oluştur
t = np.linspace(0, 4*np.pi, 75)

# Fonksiyonun başlangıç hali
y2 = psi2(1, 1, 0.)
dy2 = dpsi2(1, 1, 0.)
ddy2 = ddpsi2(1, 1, 0.)


plt.close('all')

# Etkileşimli kısmı başlat
@interact(
    A = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    w = FloatSlider(min=0, max=2, step=2./10, value=1.0, description='$\omega$'),
    phi = FloatSlider(min=-np.pi, max=np.pi, step=np.pi/20, value=0.0, description='$\phi$'),
    tv = FloatSlider(min=0, max=4*np.pi, step= np.pi/4, value=2*np.pi, description='t')
) 

def wave2(A, w, phi, tv):
    #plt.suptitle(r'$x(t) = A\;\sin(wt+\phi),\;\;\phi = \frac{%s}{20} \pi$'%int(20*phi/(np.pi)), fontsize=18, y=1.0) 
    f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True, figsize=(6,9))
    ax1.plot(t/np.pi, psi2(A, w, phi), lw=3)
    ax2.plot(t/np.pi, dpsi2(A, w, phi), lw=3)
    ax3.plot(t/np.pi, ddpsi2(A, w, phi), lw=3)
    ax1.plot(t/np.pi, y2, 'r--', lw=3)
    ax2.plot(t/np.pi, dy2, 'r--', lw=3)
    ax3.plot(t/np.pi, ddy2, 'r--', lw=3)
    ax1.text(4.25, 0, r'$x(t) = A\;\cos(wt+\phi)$', fontsize=24, horizontalalignment='left', verticalalignment='center')
    ax2.text(4.25, 0, r'$v(t) = \frac{dx}{dt} = -\omega A\;\sin(wt+\phi)$', fontsize=24, horizontalalignment='left', verticalalignment='center')
    ax3.text(4.25, 0, r'$a(t)  = \frac{dv}{dt} = -\omega^2 A\;\cos(wt+\phi)$', fontsize=24, horizontalalignment='left', verticalalignment='center')
    # formulas
    # shared axes
    ax1.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
    ax1.xaxis.set_major_locator(tck.MultipleLocator(base=.5))
    f.subplots_adjust(hspace=0)
    plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
    # set y limit
    plt.ylim(-2,2)
    # set labels
    plt.xlabel('t', fontsize=18)    
    ax1.set_ylabel('x(t)', fontsize=18)
    ax2.set_ylabel('v(t)', fontsize=18)
    ax3.set_ylabel('a(t)', fontsize=18)
    # draw y=0 or x-coordinate line
    ax1.axhline(color='black')
    ax2.axhline(color='black')
    ax3.axhline(color='black')
    ax1.axvline(x=tv/np.pi, color='orange', lw=2)
    ax2.axvline(x=tv/np.pi, color='orange', lw=2)
    ax3.axvline(x=tv/np.pi, color='orange', lw=2)
    # draw grids
    ax1.grid()
    ax2.grid()
    ax3.grid()
    plt.show()
```

```python
#*************************************************************
# Gerekli kütüphaneler en üstteki Python hücresinde yüklendi.
#*************************************************************

# Harmonik salınım fonksiyonu
def psiXY(A, w, phi, t):
    x = A * np.cos(w*t + phi)
    y = A * np.sin(w*t + phi)
    return [x, y]

# Düzgün Dairesel hareketi temsil eden daire
def daire(A, w, phi):
    x = A * np.cos(w*t + phi)
    y = A * np.sin(w*t + phi)
    return x, y

# Zaman uzayını oluştur
t = np.linspace(0, 4*np.pi, 75)
merkez = [0], [0] # dairenin merkezi

plt.close('all')

# Etkileşimli kısmı başlat
@interact(
    A = FloatSlider(min=0, max=2, step=2./10, value=1.0),
    w = FloatSlider(min=0, max=2, step=2./10, value=1.0, description='$\omega$'),
    phi = FloatSlider(min=-np.pi, max=np.pi, step=np.pi/20, value=0.0, description='$\phi$'),
    theta = FloatSlider(min=0, max=4*np.pi, step= np.pi/16, value=9*np.pi/4, description='$\\theta$')
) 

def dairesel_hareket(A, w, phi, theta):
    f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(20, 6), gridspec_kw = {'width_ratios':[1, 2]})
    f.subplots_adjust(wspace=0)
    t = np.linspace(0, theta, 75)
    
    arrowX, arrowY = A*np.cos(theta), A*np.sin(theta)
    ax1.plot(daire(A, w, phi)[0],daire(A, w, phi)[1])
    ax1.plot([psiXY(A,1,0,theta)[0], psiXY(A,1,0,theta)[0]], [0, psiXY(A,1,0,theta)[1]], color='orange')
    ax1.plot([0, psiXY(A,1,0,theta)[0]], [psiXY(A,1,0,theta)[1], psiXY(A,1,0,theta)[1]], color='orange')
    #ax1.plot(psiXY(A,1,0,theta)[0], psiXY(A,1,0,theta)[1], 'ro')
    
    ax1.quiver(merkez[0], merkez[1], [arrowX],[arrowY], angles='xy', scale_units='xy', scale=1, color='red')
    ax2.plot(t/np.pi, psiXY(A,1,0,t)[0], lw=3, label = r'x($\theta$)')
    ax2.plot(t/np.pi, psiXY(A,1,0,t)[1], lw=3, color='green', label = r'y($\theta$)')
    
    ax2.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
    ax2.xaxis.set_major_locator(tck.MultipleLocator(base=.5))
    # set y limit
    ax1.set_xlim(-2,2)
    ax1.set_ylim(-2,2)
    ax2.set_xlim(0,4)
    # set labels
    ax1.set_xlabel(r'x($\theta$)', fontsize=20)
    ax1.set_ylabel(r'y($\theta$)', fontsize=20)
    ax2.set_xlabel('$\\theta$', fontsize=20)   
    ax2.set_ylabel(r'x($\theta$), y($\theta$)', fontsize=20)
    ax2.yaxis.set_ticks_position('right')
    ax2.yaxis.set_label_position("right")
    # draw y=0 or x-coordinate line
    ax1.axhline(color='black')
    ax1.axvline(color='black')
    ax2.axhline(color='black')
    # draw grids
    ax1.grid()
    ax2.grid()
    ax1.set_aspect('equal')
    plt.legend()
    plt.show()
```

<br>    
\begin{equation}
\label{eq:ipCisim1}
\large
\vec F_{g\theta} = -m g \sin(\theta) \hat \theta
\end{equation}
<br>
halini alacaktır. Sistem başlangıçta durgun halden serbest bırakıldığında $\hat \theta$ doğrultusunda üzerindeki tek kuvvet $\vec F_{g\theta}$ olacaktır. Böylece Newton'un ikinci yasası göz önünde bulundurulursa,
<br>    
\begin{equation}
\label{eq:ipCisim2}
\large
\vec F_{g\theta} = m \vec a_{\theta}
\end{equation}
<br>    
olur. Denk. \ref{eq:ipCisim1} ve \ref{eq:ipCisim2} beraber incelenirse,
<br>    
\begin{equation}
\label{eq:ipCisim3}
\large
\vec a_{\theta} + g \sin(\theta) \hat \theta = 0
\end{equation}
<br>
 ifadesine ulaşılır. Bu denklemde $\vec a_\theta$ ve $\hat \theta$ aynı doğrultuda olduklarından,
<br>    
\begin{equation}
\label{eq:ipCisim4}
\large
a_{\theta} + g \sin(\theta) = 0
\end{equation}
<br>
yazılabilir. Burada $a_\theta$ ve $\sin(\theta)$ tek boyutlu ve zıt yönlü iki vektör gibi düşünülebilir. $a_\theta$ salınan cismin izlediği yolun zamana göre ikinci türevinin alınmasıyla bulunabilir. Herhangi bir andaki $\theta$ açısı için alınan yol şekildeki sistem için $L \theta$ kadar olacaktır. Bu ifadenin ikinci türevi $a_\theta$ ivmesini verir.
<br>    
\begin{equation}
\label{eq:ipCisim5}
\large
a_{\theta} = \frac{d^2 (L\theta)}{dt^2} = L\frac{d^2 (\theta)}{dt^2} = L \ddot{\theta} 
\end{equation}
<br>    
Böylece, birazcık düzenlemeyle, Denk. \ref{eq:ipCisim4},
\begin{equation}
\label{eq:ipCisim6}

\ddot{\theta} + {g \over L} \sin(\theta) = 0 \\
\large \ddot{\theta} + \omega^2 \sin(\theta) = 0
\end{equation}

\large

\end{equation}

<br><br>
\begin{equation}
\label{eq:ipCisim8}
\large
\sin(\theta) \eqsim \theta
\end{equation}

 <br><br>
\begin{equation}
\label{eq:ipCisim9}
\large \ddot{\theta} + \omega^2 \theta = 0
\end{equation}

\begin{equation}

### Kütle ve iki yay sistemi: enine salınım

<table width="80%" height="100%" border="1" class='b'>
<tr>
<td class='b'>
    <img width='70%' src='./sekiller/1kutle_2yay_enine.png'/>
    Kütle - iki yay sistemi [8].
</td>   
<td width='50%' class='b'> 
<br>    
Kütle ve iki yay sisteminin enine salınımlarını incelemek için yandaki şekli anlamakta fayda var. $\Large \text{O}\kern{-8.5pt} \raise 1.5pt \small \text{A}$ durumunda yaylar kütleye bağlanmamış ve serbest haldedirler. Yayların serbest uzunluklar $x_0$'dır. $\Large \text{O}\kern{-8.5pt} \raise 1.5pt \small \text{B}$ durumunda yaylar eşit miktarda gerilmiş ($x-x_0>0$) ve kütleye bağlanmıştır. Bu durumda kütle her iki taraftan birbirine zıt yöndeki $F_0$ büyüklüğündeki kuvvetlerle çekilmektedir. Böylece kütlenin üzerindeki toplam kuvvet sıfırdır. $F_0$'ın değeri,
<br><br>
\begin{eqnarray}
\label{eq:Enine2Yay1}
\large F_{0} &=& \large k(x-x_0) \nonumber 
\end{eqnarray}
<br>
dir. $\Large \text{O}\kern{-8.5pt} \raise 1.5pt \small \text{C}$ durumunda ise kütle $x=0$ çizgisinden ayrılmayacak şekilde $y$'kadar denge konumundan uzaklaştırılmıştır. Bu durumda herhangi bir anda yayların uzunluğu $L$'dir.
<br><br>
<li>Kütle ve iki yay sistemi sürtünmesiz bir şekilde $y$ doğrultusundan ayrılmayacak şekilde ayarlanmıştır.</li>
<li>Yaylar özdeştir.</li>
<li>Yaylar kütlesizdir.</li>
<li>Yaylar tarafından uygulanan kuvvetlerin iki dik bileşeni vardır.</li>
<li>$x$ bileşenleri eşit büyüklükte zıt yönlü olduklarından toplamları sıfırdır.</li>
<li>$y$ bileşenleri aynı yöndedirler ve $y$ doğrultusundaki salınım hareketini sağlarlar.</li>
</td>
</tr>
</table>

<table width="80%" height="100%" border="1" class='b'>
<td class='b'> 
Bu bilgilere göre her iki yayın kütleye uyguladığı kuvvetler;
<br><br>
\begin{eqnarray}
\label{eq:Enine2Yay2}
\large F_{S1} &=& \large F_{S2} &=& \large k (L-x_0) \nonumber \\ 
\large F_{S1,y} &=& \large F_{S2,y} &=& \large k (L-x_0) \sin{\theta} \\ 
\large \vec F_{S1,y} &=& \large \vec F_{S2,y} &=& \large - k (L-x_0)  \sin{\theta} \; \hat j\nonumber \\ 
\end{eqnarray}
<br>
olur. Şekilden anlaşılacağı üzere $\sin\theta = y/L$'dir. Yukarıdaki denklemler ve bu bilgiyle kütlenin üzerindeki $y$ doğrultusundaki toplam kuvvet aşağıdaki gibi olacaktır.
<br><br>
\begin{eqnarray}
\label{eq:Enine2Yay3}
\large \vec F_{yay, y} &=& \large \vec F_{S1,y} + \vec F_{S2,y} \nonumber \\ 
\large \vec F_{yay, y} &=& \large (- k (L-x_0) {y \over L}\hat j) + (- k (L-x_0) {y \over L}\hat j) \\
\large \vec F_{yay, y} &=& \large - 2 k (L-x_0) {y \over L}\hat j  \nonumber
\end{eqnarray}
<br>
olacaktır. Newtonun ikinci yasası uygulanırsa,

\begin{eqnarray}
\label{eq:Enine2Yay4}
\large \Sigma \vec F &=& \large m \vec a_y \nonumber \\ 
\large \Sigma \vec F &=& \large \vec F_{yay, y}  \\
\large m \vec a &=& \large  - 2 k (L-x_0) {y \over L}\hat j   \nonumber
\end{eqnarray}
<br><br>
elde edilir. Biraz matematiksel düzenleme yapılarak,
\begin{equation}
\label{eq:Enine2Yay5}
\large
\ddot{y} + {2 k \over m}(1-{x_0 \over L}) y = 0 \\
\end{equation}
<br>
differansiyel denklemine ulaşılır. Denklemden anlaşılmaktadır ki, kütle iki yay sistemi bu haliyle bir BHH değildir. İki farklı yaklaşıklık durumunda bu sistemin enine salınımları BHH gibi davranır.
</td>
</table>

#### Gevşek yay yaklaşıklığı
<table width="80%" height="100%" border="1" class='b'>
<td class='b'> 
Bu yaklaşıklığın anlamı yaylar çok gerilse bile Hooke yasasına uyma özelliklerini kaybetmemektidirler. Bu durumda $L>>x>x_0$'dır. Böylece, Denk. \ref{eq:Enine2Yay5}'teki,
<br>
\begin{equation}
\label{eq:Enine2Yay6}
\large
(1-{x_0 \over L}) \eqsim 1 \\
\end{equation}
kısım yaklaşık bir olacaktır. Bu durumda Denk. \ref{eq:Enine2Yay5} aşağıdaki hali alır.

\begin{equation}
\label{eq:Enine2Yay7}
\large
\ddot{y} + {2 k \over m} y \eqsim 0 \\
\end{equation}
Bu yaklaşıklık için sistem bir BHH gibi davranır ve boyuna salınımlarda olduğu gibi açısal frekansı $\omega = \sqrt{2k \over m}$'dir.
</td>
</table>


#### Küçük açı yaklaşıklığı
<table width="80%" height="100%" border="1" class='b'>
<td class='b'> 
Bu sistemde,    
\begin{equation}
\label{eq:Enine2Yay8}
\large
y = L \sin \theta
\end{equation}
olduğu hatırlanırsa, basit sarkaçta olduğu gibi, küçük açı yaklaşıklığın bir sonucu olarak $\sin\theta\eqsim\theta$ olacaktır.
Bu durumda, $L\eqsim x$ olacağı anlaşılabilir. Böylece,
\begin{equation}
\label{eq:Enine2Yay9}
\large
y \eqsim x\;\theta
\end{equation}    
olur. $x$ $\Large \text{O}\kern{-8.5pt} \raise 1.5pt \small \text{B}$ durumunda yayların uzunluklarıdır ve sabittir. Bu yaklaşıklık Denk. \ref{eq:Enine2Yay5}'e uygulanırsa,
\begin{equation}
\label{eq:Enine2Yay10}
\large \ddot{y} + {2 k (x-x_0) \over m x} y \eqsim 0 \\
\large \ddot{y} + {2 F_0 \over m x} y \eqsim 0 \\
\end{equation}
elde edilir. Burada sistemin açısal frekansı $\omega=\sqrt{2 F_0 \over m x}$'dir ve değeri sabittir. Böylece bu yaklaşıklık için de sistem BHH gibi davranır. <b>Bu yaklaşıklıkta ortaya çıkan açısal frekans, boyuna salınımın ve gevşek yay yaklaşımının açısal frekansından farklıdır.</b>
</td>
</table>
